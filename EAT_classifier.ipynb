# One Slice segmentation

### Data Loading and Initial Preparation

#### Confusion Matrix + ROC Curve for particular models

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.metrics import (
    confusion_matrix, roc_curve, roc_auc_score,
    ConfusionMatrixDisplay, RocCurveDisplay
)

plik = "data/data_rv_lv_eat.xlsx"
df = pd.read_excel(plik, header=None)

feature_names = df.iloc[2:, 0].values
labels = df.iloc[1, 1:].astype(int).values
data = df.iloc[2:, 1:].astype(float).values
patients = df.iloc[0, 1:].values

df_full = pd.DataFrame(data.T, index=patients, columns=feature_names)
df_full["label"] = labels

prefixes = ["eat_", "lv_", "rv_"]
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

model_defs = {
    "Random Forest": RandomForestClassifier(random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "SVM": SVC(probability=True, random_state=42),
    "KNN": KNeighborsClassifier()
}

param_grids = {
    "Random Forest": {"n_estimators": [50, 100, 200]},
    "Gradient Boosting": {"learning_rate": [0.01, 0.1, 0.2]},
    "Logistic Regression": {"C": [0.01, 0.1, 1, 10]},
    "SVM": {"C": [0.1, 1, 10]},
    "KNN": {"n_neighbors": [3, 5, 7, 9]}
}

results_table = []
X_to_split = df_full.iloc[:, :-1]
y_to_split = df_full["label"]
patients_ids = df_full.index
X_train_full, X_test_full, y_train, y_test, idx_train, idx_test = train_test_split(
    X_to_split, y_to_split, patients_ids, test_size=0.3, stratify=y_to_split, random_state=42
)

scaler = StandardScaler()
X_train_scaled_full = scaler.fit_transform(X_train_full)
X_test_scaled_full = scaler.transform(X_test_full)

X_train_scaled_full_df = pd.DataFrame(X_train_scaled_full, columns=X_train_full.columns)
X_test_scaled_full_df = pd.DataFrame(X_test_scaled_full, columns=X_test_full.columns)

for prefix in prefixes:
    X_train_prefix_full = X_train_scaled_full_df.loc[:, X_train_scaled_full_df.columns.str.startswith(prefix)]
    X_test_prefix_full = X_test_scaled_full_df.loc[:, X_test_scaled_full_df.columns.str.startswith(prefix)]

    print(f"\n\n\n**Analysis for prefix: {prefix.upper()}**\n")

    selector_all = SelectKBest(score_func=f_classif, k='all')
    selector_all.fit(X_train_prefix_full, y_train)
    all_feats = np.array(X_train_prefix_full.columns)[selector_all.get_support()]
    scores = np.nan_to_num(selector_all.scores_[selector_all.get_support()], nan=0.0)

    score_dict = dict(zip(all_feats, scores))
    sorted_feats = sorted(score_dict.items(), key=lambda x: -x[1])
    sorted_feat_names = [f for f, _ in sorted_feats]

    selected = []
    removed = []
    threshold = 0.9

    print(sorted_feats)
    print(sorted_feat_names)

    for i, feat in enumerate(sorted_feat_names):
        is_redundant = False
        for s in selected:
            corr_val = abs(X_train_prefix_full[feat].corr(X_train_prefix_full[s]))
            if corr_val > threshold:
                print(f"‚ö†Ô∏è Redundant pair: {feat} vs {s} (corr={corr_val:.2f})")
                removed.append(feat)
                is_redundant = True
                break
        if not is_redundant:
            selected.append(feat)
        if len(selected) == 10:
            break

    # If not enough features, continue adding non-redundant ones
    if len(selected) < 10:
        for feat in sorted_feat_names:
            if feat in selected or feat in removed:
                continue
            if all(abs(X_train_prefix_full[feat].corr(X_train_prefix_full[exist])) < threshold for exist in selected):
                selected.append(feat)
                print(f"‚ûï Added to complete 10 features: {feat}")
                if len(selected) == 10:
                    break

    print("\nüßπ Removed redundant features:")
    for f in removed:
        print("-", f)

    print("\n‚úÖ Final selected features (10):")
    for f in selected:
        print("-", f)



    results_prefix = []

    for name, model in model_defs.items():
        X_train_selected = X_train_prefix_full[selected]
        X_test_selected = X_test_prefix_full[selected]
        selected_features = selected


        # --- Grid search on selected features ---
        param_grid = param_grids.get(name, {})
        grid_search = GridSearchCV(
            model, param_grid, cv=cv, scoring="f1", n_jobs=-1
        )
        grid_search.fit(X_train_selected, y_train)

        best_model = grid_search.best_estimator_

        y_pred = best_model.predict(X_test_selected)
        y_proba = best_model.predict_proba(X_test_selected)[:, 1]

        cm = confusion_matrix(y_test, y_pred)
        fpr, tpr, _ = roc_curve(y_test, y_proba)
        auc = roc_auc_score(y_test, y_proba)

        tn, fp, fn, tp = cm.ravel()
        sensitivity = tp / (tp + fn) if (tp + fn) > 0 else 0
        specificity = tn / (tn + fp) if (tn + fp) > 0 else 0
        precision = tp / (tp + fp) if (tp + fp) > 0 else 0

        wrong_preds = [str(pid) for pid, true, pred in zip(idx_test, y_test, y_pred) if true != pred]

        results_prefix.append({
            "Model": name,
            "F1_CV": f"{grid_search.best_score_:.3f} ¬± {grid_search.cv_results_['std_test_score'][grid_search.best_index_]:.3f}",
            "Selected_Features": list(selected_features),
            "Misclassified_Patients": wrong_preds
        })

        results_table.append({
            "Prefix": prefix,
            "Model": name,
            "Model_trained": best_model,
            "Best_Params": grid_search.best_params_,
            "F1_CV": round(grid_search.best_score_, 3),
            "F1_Test": round(auc, 3),
            "Precision": round(precision, 3),
            "Sensitivity (Recall)": round(sensitivity, 3),
            "Specificity": round(specificity, 3),
            "Selected_Features": ", ".join(selected_features),
            "Misclassified_Patients": ", ".join(wrong_preds) if wrong_preds else "None"
        })

    for res in results_prefix:
        print(f"üåÄ {res['Model']} ‚Äì F1-score (5-fold CV): {res['F1_CV']}")
        print(f"Selected features: {res['Selected_Features']}")
        if res["Misclassified_Patients"]:
            print(f"‚ùå Errors in {res['Model']}: {res['Misclassified_Patients']}")
        else:
            print(f"‚úÖ No errors in {res['Model']}.")

    print("\n\n")

    # --- Drawing confusion matrices and ROC curves ---
    fig, axes = plt.subplots(len(model_defs), 2, figsize=(12, 5 * len(model_defs)))
    fig.suptitle(f"       Results for {prefix.upper()}", fontsize=18)

    for i, (name, model) in enumerate(model_defs.items()):
        X_train_selected = X_train_prefix_full[selected]
        X_test_selected = X_test_prefix_full[selected]
        selected_features = selected

        param_grid = param_grids.get(name, {})
        grid_search = GridSearchCV(model, param_grid, cv=cv, scoring="f1", n_jobs=-1)
        grid_search.fit(X_train_selected, y_train)
        best_model = grid_search.best_estimator_

        y_pred = best_model.predict(X_test_selected)
        y_proba = best_model.predict_proba(X_test_selected)[:, 1]

        cm = confusion_matrix(y_test, y_pred)
        fpr, tpr, _ = roc_curve(y_test, y_proba)

        disp_cm = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Healthy", "Sick"])
        disp_cm.plot(ax=axes[i, 0], cmap="Blues", colorbar=False)
        axes[i, 0].set_title(f"{name} ‚Äì Confusion Matrix")

        if y_proba is not None:
            fpr, tpr, _ = roc_curve(y_test, y_proba)
            auc = roc_auc_score(y_test, y_proba)
            RocCurveDisplay(fpr=fpr, tpr=tpr, estimator_name=name).plot(ax=axes[i, 1])
            axes[i, 1].set_title(f"{name} ‚Äì ROC Curve (AUC = {auc:.2f})")
        else:
            axes[i, 1].text(0.5, 0.5, "Predict_proba doesn't exist", ha='center', va='center')
            axes[i, 1].set_title(f"{name} ‚Äì ROC Curve")

    plt.tight_layout()
    plt.show()

df_results_summary = pd.DataFrame(results_table)
display(df_results_summary)
```

Most Important Features in Each Model
* Feature selection is performed only on the training set (X_train) using SelectKBest with f_classif as the scoring function.
* The top 10 features are selected based on their individual statistical association (ANOVA F-value) with the target variable.
* The test set (X_test) is transformed using the same selector fitted on X_train, without re-fitting ‚Äî to avoid data leakage.
* Each model is trained (and tuned via GridSearchCV) only on these selected features.
* Evaluation (predictions, scoring metrics) is performed on the test set with selected features, ensuring full separation between training and test data.
This guarantees that no information from the test set influences feature selection or model training ‚Äî preserving the validity of generalization.
* The chart below shows feature importance for each trained model, based only on the features selected in the training phase.
* Depending on the model type, importance is measured either using .feature_importances_ (e.g. trees) or the absolute value of .coef_ (e.g. linear models).


```python
import matplotlib.pyplot as plt
import numpy as np

def plot_feature_importance_all_models(results_table, prefix_filter=None):
    models_with_weights = []

    for row in results_table:
        model = row["Model_trained"]
        prefix = row["Prefix"]
        if prefix_filter and prefix != prefix_filter:
            continue

        selected_features = row["Selected_Features"].split(", ")

        if hasattr(model, "feature_importances_"):
            importance = model.feature_importances_
            models_with_weights.append((row["Model"], prefix, selected_features, importance))

        elif hasattr(model, "coef_"):
            importance = np.abs(model.coef_[0])
            models_with_weights.append((row["Model"], prefix, selected_features, importance))

        else:
            print(f"Model {row['Model']} does not support feature importance ‚Äì skipping.")

    n = len(models_with_weights)
    if n == 0:
        print("‚ùå None of the models provide feature importance or coefficients.")
        return

    fig, axes = plt.subplots(n, 1, figsize=(10, 4 * n))
    if n == 1:
        axes = [axes]

    for ax, (model_name, prefix, features, weights) in zip(axes, models_with_weights):
        ax.barh(features, weights)
        ax.set_title(f"{model_name} ‚Äì Feature Importance ({prefix.upper()})")
        ax.invert_yaxis()
        ax.set_xlabel("Value")

    plt.tight_layout()
    plt.show()

plot_feature_importance_all_models(results_table)
```

### Model based on F1 based on all features

The best model was selected based on its highest F1-score on the test set (F1_Test) across all evaluated models and prefixes (eat_, lv_, rv_).
F1-score was chosen as the primary metric because it balances precision and recall, which is critical in medical diagnostics.
After identifying the top-performing model, it was retrained on the full dataset corresponding to its prefix and saved for final use.

```python
import joblib
from sklearn.base import BaseEstimator, TransformerMixin

df_results_summary["CustomScore"] = (
    1 * df_results_summary["F1_Test"] +
    0 * df_results_summary["Sensitivity (Recall)"] +
    0 * df_results_summary["Specificity"]
)

# --- Selecting the best model based on CustomScore ---
best_model_row = df_results_summary.sort_values(by="CustomScore", ascending=False).iloc[0]

print(f"Best model overall:")
print(f"- Prefix: {best_model_row['Prefix']}")
print(f"- Model: {best_model_row['Model']}")
print(f"- Best Params: {best_model_row['Best_Params']}")
print(f"- F1_Test: {best_model_row['F1_Test']}")
print(f"- Sensitivity: {best_model_row['Sensitivity (Recall)']}")
print(f"- Specificity: {best_model_row['Specificity']}")
print(f"- CustomScore: {best_model_row['CustomScore']:.3f}")

final_model = best_model_row['Model_trained']
final_model.set_params(**best_model_row["Best_Params"])
selected_features_list = best_model_row['Selected_Features'].split(", ")

# --- Dummy selector to preserve sklearn pipeline interface ---
class DummySelector(BaseEstimator, TransformerMixin):
    def __init__(self, selected_features):
        self.selected_features = selected_features

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        return X[self.selected_features]

selector = DummySelector(selected_features_list)

joblib.dump(final_model, "best_model_selected.pkl")
joblib.dump(scaler, "scaler_selected.pkl")
joblib.dump(selector, "selector_selected.pkl")
joblib.dump(selected_features_list, "best_model_selected_features.pkl")

print("\nüíæ Final model saved as 'best_model_selected.pkl'.")
print("üíæ Scaler saved as 'scaler_selected.pkl'.")
print("üíæ Feature selector saved as 'selector_selected.pkl'.")
print("üíæ Selected features list saved as 'best_model_selected_features.pkl'.")
```

### Model with best combined features from EAT, LV, RV

I combined the best-performing features from separate anatomical regions (EAT, LV, RV) into a unified feature set.
Statistical feature selection (SelectKBest with ANOVA F-test) was applied to choose the top 10 most predictive features.
Final model selection was based on cross-validated F1-score and test set performance across multiple machine learning algorithms.

Feature Selection 
* Features are first ranked by statistical relevance using SelectKBest with the ANOVA F-test (f_classif).
* The top 10 features are selected based on their F-scores (computed on the training set only).
* Then, features with high mutual correlation (above 0.9) are considered redundant and removed.
* To maintain the total number of features (10), next-best non-redundant features are added from the ranking list.
* The final selected features are used for model training and evaluation.

```python
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    classification_report, confusion_matrix, roc_auc_score,
    ConfusionMatrixDisplay, RocCurveDisplay, roc_curve, f1_score
)
from sklearn.base import BaseEstimator, TransformerMixin
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import joblib

# --- Load data ---
df = pd.read_excel(plik, header=None)

feature_names = df.iloc[2:, 0].values
labels = df.iloc[1, 1:].astype(int).values
data = df.iloc[2:, 1:].astype(float).values
patients = df.iloc[0, 1:].values

df_full = pd.DataFrame(data.T, index=patients, columns=feature_names)
df_full["label"] = labels

# --- Prepare sets ---
k_best = 10
X = df_full.iloc[:, :-1]
y = df_full["label"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)

scaler = StandardScaler()
X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train), columns=X.columns, index=X_train.index)
X_test_scaled = pd.DataFrame(scaler.transform(X_test), columns=X.columns, index=X_test.index)

# --- Select top k features ---
selector_all = SelectKBest(score_func=f_classif, k='all')
selector_all.fit(X_train_scaled, y_train)
scores = selector_all.scores_
score_dict = dict(zip(X.columns, scores))
sorted_feats = sorted(score_dict.items(), key=lambda x: -x[1])
sorted_feat_names = [f for f, _ in sorted_feats[:100]]

original_selected = sorted_feat_names[:k_best]
print("Original selected features:")
for f in original_selected:
    print("-", f)

# --- Plot Feature Importance ---
top_k_scores = [(f, score_dict[f]) for f in original_selected]
top_k_scores_sorted = sorted(top_k_scores, key=lambda x: x[1], reverse=True)

plt.figure(figsize=(10, 6))
plt.barh([f for f, _ in top_k_scores_sorted], [s for _, s in top_k_scores_sorted], color='skyblue')
plt.xlabel("ANOVA F-score")
plt.title(f"Top {k_best} Features (Before Redundancy Removal)")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

# --- Remove redundant features ---
selected = []
removed = []
threshold = 0.9
corr_matrix = X_train_scaled[sorted_feat_names].corr().abs()

for feat in original_selected:
    is_redundant = any(corr_matrix[feat][other] > threshold for other in selected)
    if is_redundant:
        removed.append(feat)
    else:
        selected.append(feat)

# --- Add next best non-redundant features ---
for feat in sorted_feat_names:
    if feat in selected or feat in removed:
        continue
    if all(abs(X_train_scaled[feat].corr(X_train_scaled[sel])) < threshold for sel in selected):
        selected.append(feat)
        print(f"‚ûï Adding feature: {feat}")
    if len(selected) == k_best:
        break

print("\nRemoved redundant features:")
for f in removed:
    print("-", f)

print(f"\n‚úÖ Final selected features ({len(selected)}):")
for f in selected:
    print("-", f)

# --- Final feature data ---
X_train_selected = X_train_scaled[selected]
X_test_selected = X_test_scaled[selected]

# --- Model definitions ---
param_grids = {
    "Random Forest": {
        "model": RandomForestClassifier(random_state=42),
        "params": {"n_estimators": [50, 100, 200]}
    },
    "Gradient Boosting": {
        "model": GradientBoostingClassifier(random_state=42),
        "params": {"learning_rate": [0.01, 0.1, 0.2]}
    },
    "Logistic Regression": {
        "model": LogisticRegression(max_iter=1000, random_state=42),
        "params": {"C": [0.01, 0.1, 1, 10]}
    },
    "SVM": {
        "model": SVC(probability=True, random_state=42),
        "params": {"C": [0.1, 1, 10]}
    },
    "XGBoost": {
        "model": XGBClassifier(use_label_encoder=False, eval_metric="logloss", random_state=42),
        "params": {"n_estimators": [50, 100, 200], "learning_rate": [0.01, 0.1, 0.2]}
    },
    "KNN": {
        "model": KNeighborsClassifier(),
        "params": {"n_neighbors": [3, 5, 7, 9]}
    }
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
results = []

for name, entry in param_grids.items():
    grid = GridSearchCV(entry["model"], entry["params"], scoring="f1", cv=cv, n_jobs=-1)
    grid.fit(X_train_selected, y_train)
    best_model = grid.best_estimator_
    y_pred = best_model.predict(X_test_selected)
    y_proba = best_model.predict_proba(X_test_selected)[:, 1]

    cm = confusion_matrix(y_test, y_pred)
    tn, fp, fn, tp = cm.ravel()
    f1 = f1_score(y_test, y_pred)
    sens = tp / (tp + fn) if (tp + fn) > 0 else 0
    spec = tn / (tn + fp) if (tn + fp) > 0 else 0
    auc = roc_auc_score(y_test, y_proba)

    custom_score = 0.4 * f1 + 0.3 * sens + 0.3 * spec

    results.append({
        "name": name,
        "best_score": grid.best_score_,
        "best_estimator": best_model,
        "F1_Test": f1,
        "Sensitivity (Recall)": sens,
        "Specificity": spec,
        "AUC": auc,
        "CustomScore": custom_score
    })

df_results_summary = pd.DataFrame(results)
df_results_summary = df_results_summary.sort_values(by="CustomScore", ascending=False)

rf_model_entry = next(row for row in results if row["name"] == "Random Forest")
best_model = rf_model_entry["best_estimator"]
best_row = rf_model_entry  

print("\nSaving: Random Forest (F1 = {:.3f}, AUC = {:.3f})".format(
    rf_model_entry["F1_Test"], rf_model_entry["AUC"]
))


print("\nSelected best model by CustomScore:")
print(df_results_summary[["name", "F1_Test", "Sensitivity (Recall)", "Specificity", "CustomScore"]])


y_pred = best_model.predict(X_test_selected)
y_proba = best_model.predict_proba(X_test_selected)[:, 1]
cm = confusion_matrix(y_test, y_pred)
fpr, tpr, _ = roc_curve(y_test, y_proba)
auc = roc_auc_score(y_test, y_proba)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
fig.suptitle(f"Top 1: {best_row['name']}", fontsize=16)

# Confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Healthy", "Sick"])
disp.plot(ax=axes[0], cmap="Blues", colorbar=False)
axes[0].set_title(f"Confusion Matrix ({best_row['name']})")

# ROC curve
axes[1].plot(fpr, tpr, label=f"{best_row['name']} (AUC = {auc:.2f})")
axes[1].set_title(f"ROC Curve (AUC = {auc:.2f})")
axes[1].set_xlabel("False Positive Rate")
axes[1].set_ylabel("True Positive Rate")
axes[1].legend(loc="lower right")
axes[1].grid(True)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()


# --- Save model and components ---
joblib.dump(best_model, "model_rv_lv_eat.pkl")
joblib.dump(scaler, "scaler_rv_lv_eat.pkl")
joblib.dump(selected, "selected_features_rv_lv_eat.pkl")

# Dummy selector for compatibility
class DummySelector(BaseEstimator, TransformerMixin):
    def __init__(self, selected_features):
        self.selected_features = selected_features
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return X[self.selected_features]

selector = DummySelector(selected)
joblib.dump(selector, "selector_rv_lv_eat.pkl")

print("\nüíæ Saved:")
print("- model_rv_lv_eat.pkl")
print("- scaler_rv_lv_eat.pkl")
print("- selected_features_rv_lv_eat.pkl")
print("- selector_rv_lv_eat.pkl")

# ------------------
from sklearn.tree import plot_tree

# --- If the best model is Random Forest, plot a sample decision tree ---
if best_row["name"] == "Random Forest":
    print("\nüå≤ Plotting example decision tree from Random Forest...")

    # Extract one of the trees
    single_tree = best_model.estimators_[0]

    # Plot the tree (limiting depth for readability)
    plt.figure(figsize=(20, 10))
    plot_tree(
        single_tree,
        feature_names=selected,
        class_names=["Healthy", "Sick"],
        filled=True,
        rounded=True,
        max_depth=10,  # to keep the tree readable
        fontsize=10
    )
    plt.title("Example Decision Tree from Random Forest (First Estimator)")
    plt.tight_layout()
    plt.show()


```

```python
from scipy.stats import norm
import pandas as pd
import numpy as np

def sklearn_logistic_regression_summary(model, X, y, feature_names=None):
    if not hasattr(model, "coef_"):
        raise ValueError("Model must be a fitted LogisticRegression.")

    if feature_names is None:
        feature_names = [f"x{i}" for i in range(X.shape[1])]

    X_design = np.hstack([np.ones((X.shape[0], 1)), X])
    feature_names = ["Intercept"] + feature_names

    coefs = np.concatenate([model.intercept_, model.coef_.flatten()])

    p = model.predict_proba(X)[:, 1]
    V = np.diag(p * (1 - p))

    XtVX = X_design.T @ V @ X_design
    cov_matrix = np.linalg.inv(XtVX)

    se = np.sqrt(np.diag(cov_matrix))
    z = coefs / se
    p_values = 2 * (1 - norm.cdf(np.abs(z)))
    ci_lower = coefs - 1.96 * se
    ci_upper = coefs + 1.96 * se

    summary = pd.DataFrame({
        "Feature": feature_names,
        "Coefficient": coefs,
        "Std.Err": se,
        "z-score": z,
        "P>|z|": p_values,
        "[0.025": ci_lower,
        "0.975]": ci_upper
    })

    return summary

logreg_row = next((r for r in results if r["name"] == "Logistic Regression"), None)
if logreg_row:
    model = logreg_row["best_estimator"]
    X_used = X_train_selected.values
    y_used = y_train.values
    feature_names = X_train_selected.columns.tolist()

    summary_df = sklearn_logistic_regression_summary(model, X_used, y_used, feature_names)
    print(summary_df.to_string(index=False))
else:
    print("‚ö†Ô∏è Logistic Regression model not found.")
```

```python
import pandas as pd
import joblib
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from sklearn.metrics import ConfusionMatrixDisplay, RocCurveDisplay
import matplotlib.pyplot as plt

models_to_compare = {
    "Best Model (Selected)": "best_model_selected.pkl",
    "Model RV+LV+EAT": "model_rv_lv_eat.pkl",
    "Best Model (No Shape)": "best_model_no_shape.pkl",
    "Best Model (Shape)": "best_model_shape.pkl",
}

scaler_paths = {
    "Best Model (Selected)": "scaler_selected.pkl",
    "Model RV+LV+EAT": "scaler_rv_lv_eat.pkl",
    "Best Model (No Shape)": "scaler_best_model_no_shape.pkl",
    "Best Model (Shape)": "scaler_best_model_shape.pkl",
}

feature_paths = {
    "Best Model (Selected)": "best_model_selected_features.pkl",
    "Model RV+LV+EAT": "selected_features_rv_lv_eat.pkl",
    "Best Model (No Shape)": "features_best_model_no_shape.pkl",
    "Best Model (Shape)": "features_best_model_shape.pkl",
}

df = pd.read_excel(plik, header=None)


feature_names = df.iloc[2:, 0].values
labels = df.iloc[1, 1:].astype(int).values
data = df.iloc[2:, 1:].astype(float).values
patients = df.iloc[0, 1:].values

X_full = pd.DataFrame(data.T, index=patients, columns=feature_names)
y_full = labels


X_train_comparison, X_test_comparison, y_train_comparison, y_test_comparison, idx_train_comparison, idx_test_comparison = train_test_split(
    X_full, y_full, patients, test_size=0.3, stratify=y_full, random_state=42
)

results = []

for model_name, model_path in models_to_compare.items():
    print(f"\n\nüì¶ Testing model: {model_name}")

    model = joblib.load(model_path)
    scaler = joblib.load(scaler_paths[model_name])
    feature_names = joblib.load(feature_paths[model_name])

    X_for_scaler = X_test_comparison.copy()

    X_scaled = scaler.transform(X_for_scaler)
    X_scaled = pd.DataFrame(X_scaled, columns=X_for_scaler.columns, index=X_for_scaler.index)

    X_selected = X_scaled[feature_names]

    y_pred = model.predict(X_selected)
    y_proba = model.predict_proba(X_selected)[:, 1]

    report = classification_report(y_test_comparison, y_pred, output_dict=True)
    precision = round(report["1"]["precision"], 3)
    recall = round(report["1"]["recall"], 3)
    f1 = round(report["1"]["f1-score"], 3)
    specificity = round((confusion_matrix(y_test_comparison, y_pred)[0, 0]) / (confusion_matrix(y_test_comparison, y_pred)[0, 0] + confusion_matrix(y_test_comparison, y_pred)[0, 1]), 3)
    auc = round(roc_auc_score(y_test_comparison, y_proba), 3)

    results.append({
        "Model": model_name,
        "Precision": precision,
        "Recall (Sensitivity)": recall,
        "Specificity": specificity,
        "F1 Score": f1,
        "ROC AUC": auc
    })

    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    fig.suptitle(f"Results for {model_name}", fontsize=16)

    cm = confusion_matrix(y_test_comparison, y_pred)
    disp_cm = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Healthy", "Sick"])
    disp_cm.plot(ax=axes[0], colorbar=False)
    axes[0].set_title("Confusion Matrix")

    RocCurveDisplay.from_predictions(y_test_comparison, y_proba, name=model_name, ax=axes[1])
    axes[1].set_title(f"ROC Curve (AUC = {auc})")

    plt.tight_layout()
    plt.show()

df_results = pd.DataFrame(results)
display(df_results.sort_values(by="F1 Score", ascending=False))

epsilon = 4e-3  
plt.figure(figsize=(10, 8))
plt.title("ROC Curves ‚Äì All Compared Models")

for i, result in enumerate(results):
    model = joblib.load(models_to_compare[result["Model"]])
    scaler = joblib.load(scaler_paths[result["Model"]])
    features = joblib.load(feature_paths[result["Model"]])

    X_scaled = scaler.transform(X_test_comparison)
    X_scaled = pd.DataFrame(X_scaled, columns=X_test_comparison.columns, index=X_test_comparison.index)
    X_selected = X_scaled[features]

    y_proba = model.predict_proba(X_selected)[:, 1]
    fpr, tpr, _ = roc_curve(y_test_comparison, y_proba)
    fpr_shifted = np.clip(fpr + i * epsilon, 0, 1) 
    auc = roc_auc_score(y_test_comparison, y_proba)
    plt.step(fpr_shifted, tpr, where="post", label=f"{result['Model']} (AUC = {auc:.2f})", linewidth=2)

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend(loc="lower right")
plt.grid(True)
plt.tight_layout()
plt.show()

```
